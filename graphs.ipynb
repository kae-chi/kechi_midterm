```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
trainingSet = pd.read_csv("./data/train.csv")

# Define colors for the bars
colors = ["#66c2a5", "#fc8d62", "#8da0cb", "#e78ac3", "#a6d854"]

# Plot the bar chart for 'Score' value counts with specified colors
trainingSet['Score'].value_counts().plot(kind='bar', legend=True, alpha=0.8, color=colors)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define feature categories and corresponding weights
categories = [
    "Helpfulness", 
    "Time Pattern", 
    "User History", 
    "Product Pattern", 
    "Text Analysis", 
    "Historical Context"
]

# Relative importance weights based on feature counts
importance_weights = [
    0.137,  # Helpfulness Analysis (7 features)
    0.216,  # Time Pattern Analysis (11 features)
    0.137,  # User History Analysis (7 features)
    0.137,  # Product Pattern Analysis (7 features)
    0.314,  # Text Analysis (Combined 16 features)
    0.059   # Historical Context (3 features)
]

colors = ['#4f81bd', '#c0504d', '#9bbb59', '#8064a2', '#4bacc6', '#f79646']
# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(categories, importance_weights, color=colors)
plt.xlabel("Feature Categories (Grouped by Utility)")
plt.ylabel("Relative Importance (Nomralized)")
plt.title("Relative Importance of Feature Categories in Predictive Model")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Show the plot
plt.show()
```

```python
model_scores = {
    'KNeighborsClassifier': 40.736,
    'RandomForestClassifier': 56.357,
    'GradientBoostingClassifier': 62.004
}

# Create DataFrame
scores_df = pd.DataFrame(list(model_scores.items()), columns=['Accuracy', 'Model'])

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=scores_df, palette="Set3")
plt.title('Model Performance Comparison')
plt.xlabel('Accuracy (Percentage)')
plt.ylabel('Model')
plt.show()
```
